```python
fig = plt.figure(figsize = (20, 5))

subplot1 = fig.add_subplot(1,3,1)
plt.bar(data_stocks_adani_airport['Date'], data_stocks_adani_airport['Close'], edgecolor='blue')
subplot1.set_title('Yearwise Close Distribution')

subplot1 = fig.add_subplot(1,3,2)
plt.bar(data_stocks_adani_airport['Date'], data_stocks_adani_airport['Open'], edgecolor='blue')
subplot1.set_title('Yearwise Open Distribution')

subplot1 = fig.add_subplot(1,3,3)
plt.bar(data_stocks_adani_airport['Date'], data_stocks_adani_airport['Turnover'], edgecolor='blue')
subplot1.set_title('Yearwise TurnOver Distribution')
```

```python
fig = plt.figure(figsize = (20, 5))

subplot1 = fig.add_subplot(1,3,1)
plt.bar(data_stocks_adani_airport['Date'], data_stocks_adani_airport['High'], edgecolor='blue')
subplot1.set_title('Yearwise High Distribution')

subplot1 = fig.add_subplot(1,3,2)
plt.bar(data_stocks_adani_airport['Date'], data_stocks_adani_airport['Low'], edgecolor='blue')
subplot1.set_title('Yearwise Low Distribution')

subplot1 = fig.add_subplot(1,3,3)
plt.bar(data_stocks_adani_airport['Date'], data_stocks_adani_airport['Trades'], edgecolor='blue')
subplot1.set_title('Yearwise Trades Distribution')
```
